```python
import requests
```

```python
url="https://www.flipkart.com/offers-list/top-selection?screen=dynamic&pk=marketplace%3DFLIPKART~widgetType%3DdealCard~contentType%3DpersonalisedRecommendation%2FC6&marketplace=FLIPKART"
response = requests.get(url)

print("Status Code:", response.status_code)
print("Raw Text:", response.text)  # Check what is actually returned

try:
    data = response.json()
    print("JSON:", data)
except requests.exceptions.JSONDecodeError:
    print("⚠️ Response is not valid JSON")
```
