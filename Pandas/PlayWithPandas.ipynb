3. Data Cleaning

```python
df.isnull().sum()         # Count missing values
```

```python
df.dropna()               # Drop missing values
```

```python
df.duplicated()           # Check duplicates
```

```python
df.drop_duplicates()      # Drop duplicates
```

```python
df.replace("SALES", "SALE")  # Replace values

```

 4. Column & Row Operations

```python
df['Date'] = df['MONTH_ID'] + df['YEAR_ID']      # Add new column
```

```python
df.drop('Date', axis=1)                       # Drop column
```

```python
df.drop([0,1], axis=0)                       # Drop rows by index
```

5. Filtering & Selection

```python
# Convert to numeric, force errors to NaN
df['QUANTITYORDERED'] = pd.to_numeric(df['QUANTITYORDERED'], errors='coerce')

# Now you can filter row
filtered_df = df[df['QUANTITYORDERED'] > 2]
filtered_df.head()
                     
```

```python
df['QUANTITYORDERED'] = pd.to_numeric(df['QUANTITYORDERED'], errors='coerce')
df['ORDERNUMBER'] = pd.to_numeric(df['ORDERNUMBER'], errors='coerce')

df[(df['ORDERNUMBER'] > 10107) & (df['QUANTITYORDERED'] < 30)]         # Multiple conditions
```

6. Grouping & Aggregation


```python

df.groupby('MONTH_ID').sum()                    # Group & sum
```

```python
df['SALES'] = pd.to_numeric(df['SALES'], errors='coerce')

df.groupby(['MONTH_ID'])['SALES'].mean()       # Multiple columns
```

```python
df['MONTH_ID'].value_counts()                   # Count frequency
```

```python
df.pivot_table(values='SALES', index='CITY', columns='QUANTITYORDERED', aggfunc='sum')
```

7. Sorting

```python
df.sort_values(by='MONTH_ID', ascending=True)  # Sort by values
```

```python
df.sort_index()                            # Sort by index
```

8. Merging & Joining

```python
pd.concat([df, df])                      # Combine along rows (df1,df2)

```

```python
pd.merge(df, df, on='MONTH_ID')               # Merge on key
```

```python
df.join(df, lsuffix='_left', rsuffix='_right')  # Join by index
```

 9. Time Series & Date

```python
df['Date'] = pd.to_datetime(df['Date'])     # Convert to datetime
df['year'] = df['Date'].dt.year
df['month'] = df['Date'].dt.month_name()
```

```python

df.set_index('Date', inplace=True)          # Use date as index
```

```python
df.resample('M').sum()                      # Monthly resample
```

 10. Visualization (with Matplotlib/Seaborn)

```python

df['ORDERNUMBER'] = pd.to_numeric(df['ORDERNUMBER'], errors='coerce')
# Line chart (e.g., trend over time)
df['ORDERNUMBER'].plot(kind='line', title='Line Chart', figsize=(8,5), color='blue')
plt.xlabel('Index')
plt.ylabel('Values')
plt.grid(True)
plt.show()
```

```python
# Histogram (e.g., distribution)
df['ORDERNUMBER'].plot(kind='hist', title='Histogram', bins=20, color='green', edgecolor='black')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python

df['YEAR_ID'] = pd.to_numeric(df['YEAR_ID'], errors='coerce')
# Bar chart (e.g., comparison between categories)
df.plot(kind='bar', x='ORDERNUMBER', y='YEAR_ID', title='Bar Chart', color='orange', figsize=(10,6))
plt.xlabel('Category')
plt.ylabel('Value')
plt.grid(axis='y')
plt.show()
```

```python
# Box plot (e.g., outlier detection & spread)
df.plot(kind='box', title='Box Plot', figsize=(8,5), vert=False)
plt.grid(True)
plt.show()
```
