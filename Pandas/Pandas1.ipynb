```python
# 📌 Filtering and Conditional Queries
# Filter all sales in Yangon.


df[df["City"] == "Yangon"]
# Sales where quantity > 8 and rating > 9.


df[(df["Quantity"] > 8) & (df["Rating"] > 9)]
# All female customers who paid by cash.


df[(df["Gender"] == "Female") & (df["Payment"] == "Cash")]
# All sales for "Electronic accessories".


df[df["Product line"] == "Electronic accessories"]
```

```python
#  Aggregations & Grouping
# Average gross income by gender.


df.groupby("Gender")["gross income"].mean()
# Total sales per payment method.


df.groupby("Payment")["Total"].sum()
# Number of customers by type.


df["Customer type"].value_counts()
# Average tax by city.


df.groupby("City")["Tax 5%"].mean()
# Top 3 product lines by revenue.


df.groupby("Product line")["Total"].sum().sort_values(ascending=False).head(3)
```

```python
 # Data Cleaning
# Drop duplicate rows (if any).

data=df.drop_duplicates(inplace=True)
print(data)
```

```python
# Find Outliers in Rating using Box Plot


df["Rating"].plot.box()
plt.title("Boxplot of Rating")
plt.ylabel("Rating")
plt.show()
```

```python
# Correlation Matrix

correlation_matrix = df.corr(numeric_only=True)
print("Correlation matrix:\n", correlation_matrix)
```

```python
# Which Branch Has the Highest Average Total Bill?

highest_avg_branch = df.groupby("Branch")["Total"].mean().idxmax()
print("Branch with highest average bill:", highest_avg_branch)
```

```python
# Average Unit Price per Product Line

avg_price_per_product = df.groupby("Product line")["Unit price"].mean()
print("Average Unit Price per Product Line:\n", avg_price_per_product)
```
